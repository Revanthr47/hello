### Data Cleaning
#### Fixing Missing values
removing columns with null value and unnessary columns 

```python
#Finding null values in Data frame 
db.isnull().sum(axis=0)
```

```python
# % of missing datas
print(round(100*(db.isnull().sum()/len(db.index)), 2))
```

```python
#removing columns with 60% of the null values
db1 = db.dropna(axis=1,thresh=db.shape[0]*0.6,how='all')
```

```python
db1.shape
```

```python
db1.nunique().sort_values()
```

```python
#dropping the columns which has unique values as 1, as it will not affect the analysis 
db1 = db1.drop(['tax_liens','chargeoff_within_12_mths','acc_now_delinq','delinq_amnt','application_type','policy_code','initial_list_status','collections_12_mths_ex_med','pymnt_plan'],axis=1)
```

```python
db1.shape
```

Removing rows with missing values

```python
# inspecting the columns with null values  
perc_missing = round(100*(db1.isnull().sum()/len(db1.index)), 2)
perc_missing[perc_missing !=0]
```

```python
#removing null values in rows 
db1=db1[~db1.emp_title.isnull()]
db1=db1[~db1.emp_length.isnull()]
db1=db1[~db1.title.isnull()]
db1=db1[~db1.revol_util.isnull()]
db1=db1[~db1.last_pymnt_d.isnull()]
db1=db1[~db1.pub_rec_bankruptcies.isnull()]
```

```python
#dropping desc(Loan description provided by the borrower) which is not gonna affect the analysis
db1=db1.drop(['desc'],axis=1)
```

```python
db1.shape
```

```python
db1.isnull().sum(axis=0)
```

Removing addtional columns which are not useful for this analysis 
- url (URL for the LC page with listing data.)
- zip_code (The first 3 numbers of the zip code provided by the borrower in the loan application.)
- id (A unique LC assigned ID for the loan listing.)
- member_id (A unique LC assigned Id for the borrower member.)

```python
db1=db1.drop(['url','zip_code','id','member_id'],axis=1)
```

#### standardizing data

```python
db1.info()
```

```python
#changing Dtype from object to datetime for columns issue_d, last_credit_pull_d, last_pymnt_d and earliest_cr_line
db1.issue_d = pd.to_datetime(db1.issue_d, format='%b-%y')
db1.last_credit_pull_d = pd.to_datetime(db1.issue_d, format='%b-%y')
db1.last_pymnt_d = pd.to_datetime(db1.issue_d, format='%b-%y')
db1.earliest_cr_line = pd.to_datetime(db1.earliest_cr_line, format='%b-%y')
```

```python
#changing Dtype from object to float for columns that represent percerntage value - int_rate and revol_util
db1.int_rate = db1.int_rate.apply(lambda x:(x[:-1])).astype('float')
db1.revol_util = db1.revol_util.apply(lambda x:(x[:-1])).astype('float')
```
